```python
import pandas as pd
```

```python
nomefile = "test_MNIST_100_righe.txt"
with open(nomefile, "r") as fin:
    buf = fin.readlines()
    
```

```python
# costruisco il dataframe da "alimentare" con i dati del file
df = pd.DataFrame(columns=["immagine", "classe"])

for l in buf:
    vect = [x for x in l.split()[:-1] ]
    label = l.split()[-1]
    df = df.append({ "immagine": vect, "classe":  label}, ignore_index=True)
```
